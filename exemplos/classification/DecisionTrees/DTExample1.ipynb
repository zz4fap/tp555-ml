```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction import DictVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import graphviz
```

```python
# Read the CSV file
data = pd.read_csv('./decision_trees2.csv')

data
```

```python
cols_to_retain = ['Outlook', 'Temperature', 'Humidity', 'Windy']

X = data[cols_to_retain]
```

```python
y = data['Class']
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf.fit(X,y)
```

```python
dot_data = tree.export_graphviz(clf, class_names=['0', '1'], filled=True, rounded=True, special_characters=True)  
graph = graphviz.Source(dot_data)
graph
```
