```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100000

x = np.linspace(-10.0, 10.0, N)

g = 1.0 + 1.0*x

h = 1.0 / (1.0 + np.exp(-1.0*g))

yp = -1.0*np.log(h)

yn = -1.0*np.log(1-h)
```

```python
plt.plot(g, h)
plt.grid()
plt.yticks([0.0, 0.5, 1.0])
plt.xlim([-7.5, 7.5])
plt.show()
```

```python
plt.plot(g, yp, label='Classe + (y = 1)')
plt.plot(g, yn, label='Classe - (y = 0)')
plt.legend()
plt.grid()
plt.xlabel('$g(x)$')
plt.ylabel('Erro')
plt.xlim([-7.5, 7.5])
plt.ylim([-0.5, 8.0])
plt.show()
```
