```python
import sys
import sklearn
import numpy as np
from numpy.linalg import svd
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
# Reset PN-sequence generator.
seed = 4
np.random.seed(seed)
```

```python
N = 2000

x = np.linspace(0,20.1,N)

z1 = np.sin(x)
z2 = 2*np.mod(x/10,1) - 1
z3 = -np.sign(np.sin(np.pi*x/4))*0.5

plt.plot(z1)
plt.plot(z2)
plt.plot(z3)
plt.show()
```

```python
Z = np.c_[z1.reshape(N,1), z2.reshape(N,1), z3.reshape(N,1)]

A = np.array([[1.170, -0.029, 0.089],[-0.071, 1.115, -0.135],[-0.165, 0.137, 0.806]])

XT = A.dot(Z.T)

X = XT.T
```

```python
plt.plot(X[:,0])
plt.plot(X[:,1])
plt.plot(X[:,2])
plt.show()
```

```python
U, S, Vh = svd(X)

W = np.diag(1/S).dot(Vh)

YT = W.dot(X.T)

Y = YT.T

print(Y.shape)
```

```python
plt.plot(Y[:,0])
plt.plot(Y[:,1])
plt.plot(Y[:,2])
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))
ax1.plot(Y[:,0])
ax2.plot(Y[:,1])
ax3.plot(Y[:,2])
plt.show()
```
