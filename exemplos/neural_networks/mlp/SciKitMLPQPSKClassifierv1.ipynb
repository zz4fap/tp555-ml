```python
from sklearn.neural_network import MLPClassifier
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
```

```python
# Reset the PN-sequence generator.
seed = 1234
np.random.seed(seed)
```

### Define some useful functions.

```python
# Definition of several utility functions.
def mod(V):
    '''Modulador QPSK.'''
    ip = np.zeros((len(V),1),dtype=complex)
    inc = 0
    for v in V:
        if(v==0):
            ip[inc] = -1.0 - 1j*1.0
        elif(v==1):
            ip[inc] = -1.0 + 1j*1.0
        elif(v==2):
            ip[inc] = 1.0 - 1j*1.0
        else:
            ip[inc] = 1.0 + 1j*1.0
        inc += 1
    # Normalization of energy to 1.
    s = (1/np.sqrt(2))*ip;
    return s

def optimumDemod(y):
    '''Demodulador ótimo.'''
    d = np.zeros((len(y),))
    for i in range(0,len(y)):
        if(y[i, 0] >= 0 and y[i, 1] >= 0):
            d[i] = 3
        elif(y[i, 0] < 0 and y[i, 1] >= 0):
            d[i] = 1
        elif(y[i, 0] < 0 and y[i, 1] < 0):
            d[i] = 0
        elif(y[i, 0] >= 0 and y[i, 1] < 0):
            d[i] = 2
        else:
            print('Error....')
    return d

def toOneHotEncoding(V):
    '''Convert to one hot coding.'''
    bits = np.zeros((len(V),4))
    for i in range(0,len(V)):
        if(V[i]==0):
            bits[i,:] = np.array([1,0,0,0])
        elif(V[i]==1):
            bits[i,:] = np.array([0,1,0,0])
        elif(V[i]==2):
            bits[i,:] = np.array([0,0,1,0])
        else:
            bits[i,:] = np.array([0,0,0,1])
    return bits
    
def fromOneHotEncoding(bits):
    '''Convert from one hot encoding.'''
    v = np.argmax(bits, axis=0)
    return v

def detection(clf, XY, isMLP):
    '''Encapsulates both detectors into one.'''
    if(isMLP==False):
        ZZ = clf.predict(XY)
        Z = np.zeros((len(ZZ),))
        for i in range(0,len(Z)):
            Z[i] = fromOneHotEncoding(ZZ[i,:])
    else:
        Z = optimumDemod(XY)
    return Z

def plotDecisionBoundaries(Y, clf, b_train, SER, titleStr):
    '''Draw the decision boundaries.'''
    # Step size in the mesh.
    h = .02 

    plt.figure(figsize=(15, 5))
    # Plot the decision boundary. For that, we will assign a color to each point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = Y[:, 0].min() - 1, Y[:, 0].max() + 1
    y_min, y_max = Y[:, 1].min() - 1, Y[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    XY = np.c_[xx.ravel(), yy.ravel()]
    
    for i in range(0,2):
        ax = plt.subplot(1, 2, i + 1)
    
        Z = detection(clf, XY, i)
        # Put the result into a color plot.
        Z = Z.reshape(xx.shape)
        plt.pcolormesh(xx, yy, Z, shading='auto')

        # Plot also the training points.
        plt.scatter(Y[:, 0], Y[:, 1], c=b_train.ravel(), edgecolor='k', s=30)
        plt.xlim(-2.0, 2.0)
        plt.ylim(-2.0, 2.0)
        plt.xlabel('In-Phase', fontsize=14)
        plt.ylabel('Quadrature', fontsize=14)
        plt.title(titleStr[i]+' - SER: '+str(SER[i]))
    
    plt.savefig('qpsk_detection.png', dpi=600)
    plt.show()
```

### Modulation

```python
# Number of QPSK symbols to be transmitted.
N = 10000
```

```python
# Generate N binary symbols drawn from integer uniform distribution.
bits = np.random.randint(0, 4, (N, 1))    

# Modulate the binary stream into N QPSK symbols.
s = mod(bits)
```

### AWGN Channel

+ $s$ têm energia unitária, i.e., $E_s=1$.
+ $N_0 = 0.2$, onde $N_0$ é a densidade espectral do ruído.
+ Assim
$$\frac{E_s}{N_0}= -10\log_{10}(0.2) = 7 \;\;\text{[dB]}$$

```python
# Generate unit variance complex-noise vector.
noise = np.sqrt(1.0/2.0)*(np.random.randn(N, 1) + 1j*np.random.randn(N, 1))    

# Pass symbols through AWGN channel.
# A noise variance of 0.2 results in approximately 7 [dB] of Es/N0 ratio as the symbols have unit energy: (-10*log10(0.2)).
y = s + np.sqrt(0.2)*noise
```

### Compare transmitted against received signals.

```python
constellation = mod(np.array([0, 1, 2, 3]))
plt.figure()
plt.plot(y.real, y.imag, '.', label='Received signal')
plt.plot(constellation.real, constellation.imag, 'rx', markersize=10, label='Constellation')
plt.xlabel('In-Phase')
plt.ylabel('Quadrature')
plt.legend()
plt.grid()
#plt.savefig('constellation.png', dpi=600)
plt.show()
```

### Create training and validation sets.

**OBS**.: As classes da biblioteca SciKit-learn não suportam sinais complexos, então dividimos o vetor de sinais recebidos em uma matriz com dois atributos, $x_1$ e $x_2$.

```python
# Split arrays into random train and test subsets.
s_test, s_train, y_test, y_train, b_test, b_train = train_test_split(s, y, bits, random_state=seed)

# SciKit-learn's MLPs do not support complex signals, then we split the received vector into a matrix with two attributes, x1 and x2.
Y = np.c_[y_train.real, y_train.imag]
```

### Demodulation with MLP

```python
# Instantiate Multi layer Perceptron Classifier.
clf = MLPClassifier(hidden_layer_sizes=(10, 4), activation='logistic', solver='sgd', batch_size=50, learning_rate='adaptive', random_state=seed, max_iter=4000)

# Fit the MLP model.
# As it is a classifier, we have to convert the integer bits into one-hot encoding.
# The classifier will output the bit's probability.
clf.fit(Y, toOneHotEncoding(b_train))

# Prediction (detection) with trained MLP.
detected_mlp = clf.predict(np.c_[y_test.real, y_test.imag])
```

#### Accuracy

```python
acc = accuracy_score(b_test.ravel(), np.argmax(detected_mlp, axis=1))

print('Accuracy:', acc*100)
```

####  Confusion Matrix

```python
# Plot the confusion matrix
mat = confusion_matrix(b_test.ravel(), np.argmax(detected_mlp, axis=1))
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False, xticklabels=['0','1','2','3'], yticklabels=['0','1','2','3'], cmap="Blues")
plt.xlabel('true label')
plt.ylabel('predicted label')
plt.show()
```

### Demodulation with optimum detector.

Não existe detector melhor do que este no sentido da minimização da taxa de erro de símbolo (SER).

```python
# Detection with optimum detector.
detected_opt = optimumDemod(np.c_[y_test.real, y_test.imag])
```

### Calculate the SER for both detectors.

```python
# Calculate QPSK SER with MLP detector.
error = 0
for i in range(0, len(b_test)):
    if((b_test[i, 0] != fromOneHotEncoding(detected_mlp[i,:])) == True): 
        error = error + 1
    
ser_simu_mlp = 1.0 * error / len(b_test)

print('MLP SER: ', ser_simu_mlp)

# Calculate QPSK SER with optimum detector.
error = 0
for i in range(0, len(b_test)):
    if((b_test[i,0] != detected_opt[i] ) == True): 
        error = error + 1
    
ser_simu_opt = 1.0 * error / len(b_test)
    
print('OPT SER: ',ser_simu_opt)
```

```python
titleStr = ['MLP Detection', 'Optimum detection']
SER = [ser_simu_mlp, ser_simu_opt]
Y = np.c_[y.real, y.imag]
plotDecisionBoundaries(Y, clf, bits, SER, titleStr)
```

### Qual seria a vantagem em se utilizar um detector baseado em MLP?

+ Se existe um algoritmo ótimo conhecido, vindo da teoria de detecção de sinais, uma rede neural bem treinada nunca poderá superá-lo, apenas igualá-lo.



+ A detecção de sinais em canais AWGN é fácil de ser executada de forma otimizada, portanto, faz pouco sentido utilizar redes neurais artificiais para esse propósito. 



+ Existem muitas tarefas semelhantes em comunicações digitais em que redes neurais não trazem melhorias significativas.
    + Codificadores de canal.
    + Alocação de potência em redes celulares.
    + Processamento de sinais em sistemas com múltiplas antenas (MIMO).
    

No entanto, existem alguns papéis importantes que redes neurais podem desempenhar na área de comunicações digitais:

* Em primeiro lugar, existem muitos problemas em que um **algoritmo de otimização encontra uma solução ótima, mas tem uma complexidade proibitivamente alta para implementação em tempo real**. 



* Em segundo lugar, existem casos em que os **modelos de um sistema são inadequados ou incompletos**, como por exemplo amplificadores não lineares, canais ópticos não-lineares, etc.
