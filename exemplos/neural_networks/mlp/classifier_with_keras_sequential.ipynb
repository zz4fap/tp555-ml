```python
def createLogDir(root_logdir="tf_logs"):
    now = datetime.utcnow().strftime("%Y%m%d%H%M%S")
    logdir = "{}/run-{}/".format(root_logdir, now)
    return logdir
```

```python
logDir = createLogDir(root_logdir="sequential_classifier")
print(logDir)
```
