```python
# Import all the necessary libraries.
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
# Reset pseudo random number generator so that the results are reproducible.
np.random.seed(42)
```

### Generating the target function and its noisy version.

Neste exemplo, treinamos polinômios de ordem 1, 5, 10, e 20 para se ajustarem à seguinte função objetivo

$$y = 1 + 0.5 x.$$

Geramos 20 amostras da função objetivo e as corrompemos com ruído Gaussiano normal padrão.

```python
# Sets the number of examples.
N = 20

# Attribute.
x = np.sort(3*np.random.rand(N, 1), axis=0)

# Target function.
y = 1 + 0.5*x

# Noise.
w = np.random.randn(N, 1)

# Observable function (noisy).
y_noisy = y + w
```

```python
# Plot original data versus noisy version.
plt.figure()
plt.plot(x, y, label='Função objetivo')
plt.plot(x, y_noisy, 'x', label='Função ruidosa')
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.legend()
plt.grid()
#plt.savefig("target_function.png", dpi=600)
#Show the plot.
plt.show()
```

### Polinômio de ordem 1

```python
# Polynomial order.
model_order = 1
```

```python
# Instantiate a polynomial features object.
poly = PolynomialFeatures(model_order, include_bias=False)

# Create the polynomial matrix.
X = poly.fit_transform(x)

# Instantiate standard scaler.
scaler = StandardScaler()

# Create the attribute matrix and standardize the samples.
X = scaler.fit_transform(X)

# Instantiate.
reg1 = LinearRegression()

# Train.
reg1.fit(X, y_noisy)

# Predict.
ypred = reg1.predict(X)

# Plot original data versus noisy version.
plt.plot(x, y, label='Função objetivo')
plt.plot(x, y_noisy, 'x', label='Função ruidosa')
plt.plot(x, ypred, label='Função hipótese')
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.legend()
plt.grid()
plt.axis([0, 3, 0, 4])
#Show the plot.
plt.show()
```

### O quão bem um modelo que se sobreajusta muito ao conjunto de treinamento generaliza?

Vamos usar o modelo de ordem 20, pois é o que memoriza todas as amostras do conjunto de treinamento, ou seja, tem o maior sobreajuste possível, pois resulta em um erro de treinamento praticamente igual a 0.

Geramos 10.000 amostras no intervalo de 0 a 3.

```python
N = 100000

xnew = np.linspace(0,3,N).reshape(N,1)

# Create the polynomial matrix.
X = poly.transform(xnew)

# Create the attribute matrix and standardize the samples.
X = scaler.transform(X)

# Predict with the regressor.
ypred = reg20.predict(X)

# Plot original data versus noisy version.
plt.plot(x, y, label='Função objetivo')
plt.plot(x, y_noisy, 'x', label='Função ruidosa')
plt.plot(xnew, ypred, label='Função hipótese')
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.legend()
plt.grid()
plt.axis([0, 3, -100, 100])
#Show the plot.
plt.show()
```

### Conclusão

+ O aumento da flexibilidade do modelo resulta no aumento da magnitude dos pesos.


+ Modelos muito flexíveis se ajustam perfeitamento às amostras do conjunto de treinamento, mas têm uma capacidade generalização muito pequena.


+ O que a **regularização faz é restringir as magnitudes** e, **consequentemente, minimizar a função de erro e a flexibilidade do modelo**.
