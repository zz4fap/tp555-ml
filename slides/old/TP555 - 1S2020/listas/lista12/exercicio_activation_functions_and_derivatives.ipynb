```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def identity(z):
    return z

def softplus(z):
    return np.log(1 + np.exp(z))

def sinc(z):
    y = np.zeros((len(z),))
    for i in range(0,len(z)):
        if(z[i] == 0.0):
            y[i] = 1
        else:
            y[i] = np.sin(z[i])/z[i]
    return y

def elliotsig(z):
    return z/(1+ np.abs(z))

def derivative(f, z, eps=0.0000001):
    return (f(z + eps) - f(z - eps))/(2 * eps)
```

```python
z = np.linspace(-5, 5, 200)

plt.figure(figsize=(11,4))

plt.subplot(121)
plt.plot(z, identity(z), "r-", linewidth=1, label="Identity")
plt.plot(z, softplus(z), "g--", linewidth=2, label="SoftPlus")
plt.plot(z, sinc(z), "b-", linewidth=2, label="Sinc")
plt.plot(z, elliotsig(z), "m-.", linewidth=2, label="ElliotSig")
plt.grid(True)
plt.legend(fontsize=14)
plt.title("Activation functions", fontsize=14)
plt.axis([-5, 5, -5, 5])

plt.subplot(122)
plt.plot(z, derivative(identity, z), "r-", linewidth=1, label="Identity")
plt.plot(z, derivative(softplus, z), "g--", linewidth=2, label="SoftPlus")
plt.plot(z, derivative(sinc, z), "b-", linewidth=2, label="Sinc")
plt.plot(z, derivative(elliotsig, z), "m-.", linewidth=2, label="ElliotSig")
plt.grid(True)
plt.legend(loc="center right", fontsize=14)
plt.title("Derivatives", fontsize=14)
plt.axis([-5, 5, -0.5, 1.1])

plt.show()
```
